```python
import pandas as pd
import numpy as np
```

### Read csv file

```python
# Columns that we are interested
cols = ['Name', 'Age', 'Photo', 'Nationality', 'Flag', 'Overall', 'Potential', 'Club', 'Club Logo', 'Value', 'Wage', 'Release Clause']
data = pd.read_csv('data/data.csv', usecols=cols)
data.head()
```

### Queries

```python
data[data['Age'] > 42]
```

```python
def get_id(id):
    return data[data['ID'] == id]
```

```python
def older_than(age):
#     timeit df.T.to_dict().values()
    return data[data['Age'] > age].T.to_dict()
```

```python
get_id(158023).to_dict()
```
